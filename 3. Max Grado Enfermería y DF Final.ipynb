# Auxiliares

```python
fecha = pd.to_datetime('2020-06-30')
anios = 10

#Generamos los periodos deseados
periodos = pd.date_range(end=fecha, periods=anios, freq=pd.DateOffset(years=1))

#Generamos las columnas deseadas
for periodo in periodos:
    #Creamos el título de la columna y la creamos con el método assign()
    col = 'aux_' + str(periodo.year)
    
    #Copiamos la fecha de fallecido en las columnas
    kwargs = {col : lambda x: ( 
        (x['fecha_aux_ok'] == x['FECHA_TITULO_OK']) & (x['fecha_aux_ok'] < periodo) &
         (((x['dummie_tec'] == False)| ( x['dummie_tec'] == True) & (x['fecha_tec_ok'] > periodo)) &
             ((x['dummie_lic'] == False)| ( x['dummie_lic'] == True) & (x['fecha_lic_ok'] > periodo)))
                              )
             } #calculo el intervalo de inactividad segun fecha de vigencia
    subset_enf_nivel_superior = subset_enf_nivel_superior.assign(**kwargs)
```

```python
#Obtenemos los datos y generamos el dataframe
#Fecha base y periodo
fecha = pd.to_datetime('2020-06-30')
anios = 10

#Generamos los periodos deseados
periodos = pd.date_range(end=fecha, periods=anios, freq=pd.DateOffset(years=1))

def age(birth_date):
    today = periodo
    y = today.year - birth_date.year
    if today.month < birth_date.month or today.month == birth_date.month and today.day < birth_date.day:
        y -= 1
    return y

#Generamos las columnas deseadas
for periodo in periodos:
    #Creamos el título de la columna y la creamos con el método assign()
    col = 'edad_' + str(periodo.year)
    #Copiamos la fecha de fallecido en las columnas
    kwargs = {col : lambda x: pd.to_datetime(x['FECHA_NACIMIENTO'])}
    df_final = df_final.assign(**kwargs)
    
    #Sustituímos la fecha de fallecido con True o False dependiendo del periodo(Fecha de corte de la columna)
    df_final[col] = df_final[col].apply(age)
    labels = ['-24', '25-29', '30-34', '35-39','40-44','45-49','50-55','54-59','60-64', '+65']
    bins = [0, 24, 29, 34, 39, 44,49,54, 59, 64, 1000]
    df_final[col] = pd.cut(df_final[col],bins, labels = labels,include_lowest = True)
```
