```python
import plotly  
import plotly.graph_objs as go
import plotly.express as px
```

## Pirámide de Médicos

```python
df_cnps_mediques = df_cnps_prof_prov[df_cnps_prof_prov['prof_adjudicada'] == 'Médico'][['prov_adjudicada','sexo', 'grupo_etario', 'total']]
```

```python
df_cnps_nac_mediques = df_cnps_mediques[['sexo', 'grupo_etario', 'total']]
```

```python
df_cnps_nac_mediques = df_cnps_nac_mediques.groupby(['grupo_etario', 'sexo']).sum().reset_index()
```

```python
df_piramide_med = df_cnps_nac_mediques.groupby(
    ['grupo_etario', 'sexo']).sum().unstack().fillna(0).reset_index().sort_values('grupo_etario', ascending = False).droplevel(0, axis = 1).rename({'':'grupo_etario'}, axis = 1)
```

```python
df_piramide_med
```

```python
trace1 = go.Bar(
    y= ['60-64','54-59', '50-55','45-49','40-44', '35-39','30-34','25-29', '-25'],
    x= df_piramide_med['M']*-1,
    name='Varones',
    orientation = 'h',
    text= df_piramide_med['M'].astype('int'),
    hoverinfo='text',
    marker = dict(
        color = 'rgba(58, 71, 80, 0.6)',
        line = dict(
            color = 'rgba(58, 71, 80, 1.0)',
            width = 3)
    )
)
trace2 = go.Bar(
    y= ['60-64','54-59', '50-55','45-49','40-44', '35-39','30-34','25-29', '-25'],
    x=df_piramide_med['F'],
    name='Mujeres',
    orientation = 'h',
    text= df_piramide_med['F'].astype('int'),    
    hoverinfo= 'x',
    marker = dict(
        color = 'rgba(30, 130, 76, 0.6)',
        line = dict(
            color = 'rgba(30, 130, 76, 1.0)',
            width = 3)
    )
)

data = [trace1, trace2]
layout = go.Layout(
    barmode='overlay',
    xaxis=dict(
        range=[-20000, 20000]
    ),
    yaxis=dict(autorange="reversed")
)

fig_med = go.Figure(data=data, layout=layout)

fig_med.update_layout(
    height=500,
    title_text='Pirámide poblacional de Médicas/os en Argentina (Fuente: REFEPS - 2020)'
)



df_piramide_med.to_excel("tablas_graficos_profesionales/df_piramide_med_2020.xlsx")
fig_med.write_image("tablas_graficos_profesionales/df_piramide_med_2020.png")

fig_med
```
